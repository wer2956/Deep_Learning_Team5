```python
folder_list = os.listdir(test_dir)
for foldername in folder_list:
    imsidir = os.path.join(test_dir, foldername)
    image_list = os.listdir(imsidir)
    for filename in image_list:
        imagename = os.path.join(imsidir,filename)
        img = Image.open(imagename)
        width, height = img.size
        # square matrix가 되도록 통일 (shape은 512 x 512)
        if width == 720:
            area = (149, 32, 661, 544)
            cropped_img = img.crop(area)
            cropped_img.save(imagename)
        if width == 1920:
            w, h = img.size
            w = int(w/2)
            h = int(h/2)
            img_r = img.resize((w,h), resample=0, box=None)
            area = (309, 12, 821, 524)
            cropped_img = img_r.crop(area)
            cropped_img.save(imagename)
        if width == 1280:
            w, h = img.size
            w = int(w/2)
            h = int(h/2)
            img_r = img.resize((w,h), resample=0, box=None)
            area = (64, 0, 576, 512)
            cropped_img = img_r.crop(area)
            cropped_img.save(imagename)
        if width == 1350:
            if height == 1080:
                w, h = img.size
                w = int(w/2)
                h = int(h/2)
                img_r = img.resize((w,h), resample=0, box=None)
                area = (81, 14, 593, 526)
                cropped_img = img_r.crop(area)
                cropped_img.save(imagename)
            else:
                w, h = img.size
                w = int(w/2)
                h = int(h/2)
                img_r = img.resize((w,h), resample=0, box=None)
                area = (81, 10, 593, 522)
                cropped_img = img_r.crop(area)
                cropped_img.save(imagename)        
        if width == 576:
            w, h = img.size
            img_r = img.resize((w,h), resample=0, box=None)
            area = (39, 39, 551, 551)
            cropped_img = img_r.crop(area)
            cropped_img.save(imagename)
        
        print(img.size)
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(train_dir,
                                                   target_size=(128,128),
                                                   batch_size=720,
                                                   class_mode='categorical')

test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(test_dir,
                                                 target_size=(128,128),
                                                 batch_size=1,
                                                 class_mode='categorical')

for data_batch, labels_batch in train_generator:
    print('배치 데이터 크기:', data_batch.shape)
    print('배치 레이블 크기:', labels_batch.shape)
    break
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(128,128,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(8, activation='softmax'))

model.summary()
```

```python
def precision(y_target, y_pred):
    # clip(t, clip_value_min, clip_value_max) : clip_value_min~clip_value_max 이외 가장자리를 깎아 낸다
    # round : 반올림한다
    y_pred_yn = K.round(K.clip(y_pred, 0, 1)) # 예측값을 0(Negative) 또는 1(Positive)로 설정한다
    y_target_yn = K.round(K.clip(y_target, 0, 1)) # 실제값을 0(Negative) 또는 1(Positive)로 설정한다

    # True Positive는 실제 값과 예측 값이 모두 1(Positive)인 경우이다
    count_true_positive = K.sum(y_target_yn * y_pred_yn) 

    # (True Positive + False Positive) = 예측 값이 1(Positive) 전체
    count_true_positive_false_positive = K.sum(y_pred_yn)

    # Precision = (True Positive) / (True Positive + False Positive)
    # K.epsilon()는 'divide by zero error' 예방차원에서 작은 수를 더한다
    precision = count_true_positive / (count_true_positive_false_positive + K.epsilon())

    # return a single tensor value
    return precision

def recall(y_target, y_pred):
    # clip(t, clip_value_min, clip_value_max) : clip_value_min~clip_value_max 이외 가장자리를 깎아 낸다
    # round : 반올림한다
    y_target_yn = K.round(K.clip(y_target, 0, 1)) # 실제값을 0(Negative) 또는 1(Positive)로 설정한다
    y_pred_yn = K.round(K.clip(y_pred, 0, 1)) # 예측값을 0(Negative) 또는 1(Positive)로 설정한다

    # True Positive는 실제 값과 예측 값이 모두 1(Positive)인 경우이다
    count_true_positive = K.sum(y_target_yn * y_pred_yn) 

    # (True Positive + False Negative) = 실제 값이 1(Positive) 전체
    count_true_positive_false_negative = K.sum(y_target_yn)

    # Recall =  (True Positive) / (True Positive + False Negative)
    # K.epsilon()는 'divide by zero error' 예방차원에서 작은 수를 더한다
    recall = count_true_positive / (count_true_positive_false_negative + K.epsilon())

    # return a single tensor value
    return recall

def f1score(y_target, y_pred):
    _recall = recall(y_target, y_pred)
    _precision = precision(y_target, y_pred)
    # K.epsilon()는 'divide by zero error' 예방차원에서 작은 수를 더한다
    _f1score = ( 2 * _recall * _precision) / (_recall + _precision+ K.epsilon())
    
    # return a single tensor value
    return _f1score
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['acc', precision, recall, f1score])

history = model.fit_generator(train_generator,
                              steps_per_epoch=20,
                              epochs=10,
                              validation_data=test_generator,
                              validation_steps=100)
```
